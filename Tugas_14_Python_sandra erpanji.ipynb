```
import numpy as np
import matplotlib.pyplot as plt
```

```
def cos(frek,tipe,ht,hr,d):
    if frek in range(150,1501):
        Ca=69.55
        Cb=26.16
    elif frek in range(1501,2001):
        Ca=46.3
        Cb=33.9
    
    if tipe=='urban' :
        Cm=0
    elif tipe=='suburban' :
        Cm= -2*((np.log10((frek/28)**2))-5.4)
    elif tipe=='open' :
        Cm= -4.78*(np.log10(frek))+18.33*(np.log10(frek))-40.94
    else :
        Cm='tipe salah'
    
    ahr = (1.1*np.log10(frek)-0.7)*hr-(1.56*np.log10(frek)-1.8)
    Lp = Ca+(Cb*np.log10(frek))-(13.83*np.log(ht))-ahr+((44.9-6.55*np.log10(ht))*np.log10(d))-Cm
    
    
    return Lp
```

```
f=1800 #MHz
loc='urban'
loc2='suburban'
loc3='open'
tower=80 #m
user=1.5 #m
dmin=300 #m
dmax=1000 #m
N=200 #titik
d=np.random.randint(dmin,dmax,N)
```

```
print(d)
```

```
LpUrban=cos(f,loc,tower,user,d)
print(LpUrban)
```

```
Ptx=30000 #mW
PtxdBm=44.77 #dBm
Gtx=10 #dB
Grx=2 #dB
Prx1=PtxdBm-LpUrban+Gtx+Grx
print(Prx1)
```

```
mWPrx1=10**(Prx1/10)
print(mWPrx1)
```

```
LpSuburban=cos(f,loc2,tower,user,d)
print(LpSuburban)
```

```
Ptx=30000 #mW
PtxdBm=44.77 #dBm
Gtx=10 #dB
Grx=2 #dB
Prx2=PtxdBm-LpSuburban+Gtx+Grx
print(Prx2)
```

```
mWPrx2=10**(Prx2/10)
print(mWPrx2)
```

```
LpOpen=cos(f,loc3,tower,user,d)
print(LpOpen)
```

```
Ptx=30000 #mW
PtxdBm=44.77 #dBm
Gtx=10 #dB
Grx=2 #dB
Prx3=PtxdBm-LpOpen+Gtx+Grx
print(Prx3)
```

```
mWPrx3=10**(Prx3/10)
print(mWPrx3)
```

```
d_sort=d[np.argsort(d)]
Prx1_sort=mWPrx1[np.argsort(d)]
Prx2_sort=mWPrx2[np.argsort(d)]
Prx3_sort=mWPrx3[np.argsort(d)]
```

```
plt.figure()
plt.plot(d_sort,Prx1_sort,'b', label='Urban')
plt.plot(d_sort,Prx2_sort,'r', label='Suburban')
plt.plot(d_sort,Prx3_sort,'g', label='Open')
plt.legend()
plt.xlabel('Jarak (m)')
plt.ylabel('Daya Terima (mW)')
plt.show()
```

```
Noise=-174+np.log10(5000000)
NoisemW=10**(Noise/10)
SNR1=mWPrx1/NoisemW
print(SNR1)
```

```
SNR2=mWPrx2/NoisemW
print(SNR2)
```

```
SNR3=mWPrx3/NoisemW
print(SNR3)
```

```
d_sort=d[np.argsort(d)]
SNR1_sort=SNR1[np.argsort(d)]
SNR2_sort=SNR2[np.argsort(d)]
SNR3_sort=SNR3[np.argsort(d)]
```

```
plt.figure()
plt.plot(d_sort,SNR1_sort,'b', label='Urban')
plt.plot(d_sort,SNR2_sort,'r', label='Suburban')
plt.plot(d_sort,SNR3_sort,'g', label='Open')
plt.legend()
plt.xlabel('Jarak (m)')
plt.ylabel('SNR (mW)')
plt.show()
```

```
Ks1=5000000*np.log2(1+SNR1)
print(Ks1)
```

```
Ks2=5000000*np.log2(1+SNR2)
print(Ks2)
```

```
Ks3=5000000*np.log2(1+SNR3)
print(Ks3)
```
